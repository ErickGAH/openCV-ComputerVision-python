```python
import cv2
import glob
import os
import matplotlib.pyplot as plt 
import numpy as np

def color(h):
    coloraux=[]
    nameaux=""   
    if h < 10:     
        coloraux=[0,0,255]
        nameaux="rojo"
    elif h < 45:
        coloraux=[0,255,255]
        nameaux="Amarillo"
    elif h < 90:
        coloraux=[0,255,0]
        nameaux="Verde"
    elif h < 120:
        coloraux=[255,0,0]
        nameaux="Azul"
    elif h < 181:
        coloraux=[0,0,255]
        nameaux="Rojo"
    return coloraux,nameaux

# Listar los archivos jpg en el directorio
images_paths = glob.glob(os.path.join("./data", "*.jpg"))
image1 = cv2.imread(images_paths[0])
image2 = cv2.imread(images_paths[1])
image3 = cv2.imread(images_paths[2])

colorD1=[]
colorD2=[]
colorD3=[]
colorD4=[]
colorD1Name=""
colorD2Name=""
colorD3Name=""
colorD4Name=""

boxes = [
    [(248, 246, 468, 432)],  #Formato [x1, y1, x2, y2]
    [(131, 162, 246, 281)],
    [(194, 419, 298, 523), (147, 300, 195, 346)]
]

image1Hsv = cv2.cvtColor(image1,cv2.COLOR_BGR2HSV)
image2Hsv = cv2.cvtColor(image2,cv2.COLOR_BGR2HSV)
image3Hsv = cv2.cvtColor(image3,cv2.COLOR_BGR2HSV)

rec1x1,rec1y1,rec1x2,rec1y2=boxes[0][0]
rec2x1,rec2y1,rec2x2,rec2y2=boxes[1][0]
rec3x1,rec3y1,rec3x2,rec3y2=boxes[2][0]
rec4x1,rec4y1,rec4x2,rec4y2=boxes[2][1]

roi1=image1Hsv[rec1y1:rec1y2,rec1x1:rec1x2]
roi2=image2Hsv[rec2y1:rec2y2,rec2x1:rec2x2]
roi3=image3Hsv[rec3y1:rec3y2,rec3x1:rec3x2]
roi4=image3Hsv[rec4y1:rec4y2,rec4x1:rec4x2]

H, S, V = cv2.split(roi1)
hist_H = cv2.calcHist([H], [0], None, [180], [0, 180])
valor_max = np.max(hist_H)
indice_color_dominante = np.argmax(hist_H)
colorD1,colorD1Name=color(indice_color_dominante)

H, S, V = cv2.split(roi2)
hist_H = cv2.calcHist([H], [0], None, [180], [0, 180])
valor_max = np.max(hist_H)
indice_color_dominante = np.argmax(hist_H)
colorD2,colorD2Name=color(indice_color_dominante)

H, S, V = cv2.split(roi3)
hist_H = cv2.calcHist([H], [0], None, [180], [0, 180])
valor_max = np.max(hist_H)
indice_color_dominante = np.argmax(hist_H)
colorD3,colorD3Name=color(indice_color_dominante)

H, S, V = cv2.split(roi4)
hist_H = cv2.calcHist([H], [0], None, [180], [0, 180])
valor_max = np.max(hist_H)
indice_color_dominante = np.argmax(hist_H)
colorD4,colorD4Name=color(indice_color_dominante)

cv2.rectangle(image1, (rec1y1,rec1x1),(rec1x2,rec1y2),colorD1, thickness = 2, lineType = cv2.LINE_8)
cv2.putText(image1, colorD1Name, (rec1y1,rec1x1), cv2.FONT_HERSHEY_SIMPLEX,
            1, colorD1, 3, cv2.LINE_AA)

cv2.rectangle(image2, (rec2x1,rec2y1),(rec2x2,rec2y2),colorD2, thickness = 2, lineType = cv2.LINE_8)
cv2.putText(image2, colorD2Name, (rec2x1,rec2y1), cv2.FONT_HERSHEY_SIMPLEX,
            1, colorD2, 3, cv2.LINE_AA)

cv2.rectangle(image3, (rec3x1,rec3y1),(rec3x2,rec3y2),colorD3, thickness = 2, lineType = cv2.LINE_8)
cv2.putText(image3, colorD3Name, (rec3x1,rec3y1), cv2.FONT_HERSHEY_SIMPLEX,
            1, colorD3, 3, cv2.LINE_AA)

cv2.rectangle(image3, (rec4x1,rec4y1),(rec4x2,rec4y2),colorD4, thickness = 2, lineType = cv2.LINE_8)
cv2.putText(image3, colorD4Name, (rec4x1,rec4y1), cv2.FONT_HERSHEY_SIMPLEX,
            1, colorD4, 3, cv2.LINE_AA)
k=0
while(True):
  # Read frame
        
    if(k==ord('q')):
        break
    cv2.imshow("imagen1",image1)
    cv2.imshow("imagen2",image2)
    cv2.imshow("imagen3",image3)      
    
    k= cv2.waitKey(1) 

cv2.destroyAllWindows()
    
```
